```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px
import numpy as np
from scipy.stats import norm
from scipy import stats
import plotly.figure_factory as ff
```

```python
train_df = pd.read_csv("Datasets/house-prices-advanced-regression-techniques/train.csv")
```

```python
saleprice_k = [train_df['SalePrice']/1000]
fig = ff.create_distplot(hist_data=saleprice_k, 
                         group_labels=['SalePrice'], 
                         show_rug=False,
                         )
fig.update_layout(xaxis_title="SalePrice, k",
                  yaxis_title="Density",
                  height = 600, 
                  width = 600,
                  font_color='#484848', 
                  template='plotly_white',
                  )
fig2 = ff.create_distplot(hist_data=saleprice_k, 
                          group_labels=['SalePrice'], 
                          curve_type = 'normal',
                          )
normal_x = fig2.data[1]['x']
normal_y = fig2.data[1]['y']
fig.add_traces(go.Scatter(x=normal_x, 
                          y=normal_y, 
                          mode = 'lines',
                          line = dict(color='black',
                                      width = 2),
                          name = 'Normal',
                          ))
fig.update_traces(marker=dict(color='#46B3B3'),
                  hovertemplate ="<b>X: %{x}<br>Y: %{y}</b>",
)
fig.show()
```

```python
corr = train_df.corr()
trace = go.Figure(go.Heatmap(z=corr.values,
                             x=corr.index.values,
                             y=corr.columns.values,
                             colorscale=px.colors.diverging.PRGn, 
                             zmin=-1,
                             zmax=1,
                             ))
trace.update_layout(height = 900, 
                    width = 900,
                    font_color='#484848',
                    )
trace.show()
```

```python
corrmat = train_df.corr()
k = 10 #number of variables for heatmap
cols = corrmat.nlargest(k, 'SalePrice')['SalePrice'].index
cm = np.corrcoef(train_df[cols].values.T)
trace = go.Figure(go.Heatmap(z=cm,
                             x=cols.values,
                             y=cols.values,
                             colorscale=px.colors.diverging.PRGn, 
                             zmin=-1,
                             zmax=1,
                             ))
trace.update_layout(height = 600, 
                    width = 600,
                    font_color='#484848',
                    )
trace.update_traces(text=cm,
                    texttemplate="%{text:.2f}",
                    hovertemplate=None,
                    )
trace.show()
```

```python
cols = ['SalePrice', 'OverallQual', 'GrLivArea', 'GarageCars', 'TotalBsmtSF', 'FullBath', 'YearBuilt']

fig = px.scatter_matrix(train_df,
    dimensions=cols,
    title="Matrix",
    ) # remove underscore
fig.update_traces(diagonal_visible=False)
fig.update_layout(height = 1500, 
                  width = 1500,
                  font_color='#484848',
                  template='plotly_white'
                  )
fig.update_traces(marker=dict(color='#46B3B3'))
fig.show()
```
